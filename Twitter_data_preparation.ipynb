Labelling the sentiments of each tweets using teransformers

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
roberta = "cardiffnlp/twitter-xlm-roberta-base-sentiment"
```

```python
model = AutoModelForSequenceClassification.from_pretrained(roberta)
tokenizer = AutoTokenizer.from_pretrained(roberta)
```

```python
def LabelData(tweet_proc):
    
    encoded_tweet = tokenizer(tweet_proc, return_tensors='pt')
    output = model(**encoded_tweet)
    
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    
    scores_dict = {'Negative' : scores[0], 'Neutral' : scores[1],'Positive' : scores[2] }
    label = max(zip(scores_dict.values(), scores_dict.keys()))[1]
    
    print(tweet_proc,'->->->->',label)
    return label

dfen['target'] = dfen['cleaned_tweets'].apply(LabelData)
dfen
```
